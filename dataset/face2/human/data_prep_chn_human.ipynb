## Human Data

```python
import pandas as pd
import numpy as np
import json
import os
```

```python
news_df = pd.read_json("news.json")
webnovel_df = pd.read_json("webnovel.json")
wiki_df = pd.read_json("wiki.json")
```

```python
def add_target_labels(df, target_label):
    df['target'] = target_label
    return df

def analyse_data(df):
    print(f"Number of rows: {len(df)}")
    print("Maximum length of text:", df['text'].str.len().max())
    print("Minimum length of text:", df['text'].str.len().min())
    print("Average length of text:", df['text'].str.len().mean())
```

```python
# Combine all dataframes
combined_df = pd.concat([news_df, webnovel_df, wiki_df], ignore_index=True)

# Drop 'input' column
combined_df = combined_df.drop(columns=['input'], errors='ignore')

# Rename 'output' column to 'text'
combined_df = combined_df.rename(columns={'output': 'text'})

# remove empty rows
combined_df = combined_df[combined_df['text'].str.strip() != '']

# add target labels
combined_df = add_target_labels(combined_df, 0)

# analyse data
analyse_data(combined_df)
```

```python
combined_df
```

### Sample 200 data from webnovel

```python
webnovel_df_200 = webnovel_df.sample(200, random_state=42)

# drop 'input' column
webnovel_df_200 = webnovel_df_200.drop(columns=['input'], errors='ignore')

# rename 'output' column to 'text'
webnovel_df_200 = webnovel_df_200.rename(columns={'output': 'text'})

# remove empty rows
webnovel_df_200 = webnovel_df_200[webnovel_df_200['text'].str.strip() != '']

# add target labels 
webnovel_df_200 = add_target_labels(webnovel_df_200, 0)

# analyse data
analyse_data(webnovel_df_200)
```
