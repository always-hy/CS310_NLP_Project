## AI-generated Data

```python
import pandas as pd
import numpy as np
import json
```

```python
news_df_generated = pd.read_json('news.json')
webnovel_df_generated = pd.read_json('webnovel.json')
wiki_df_generated = pd.read_json('wiki.json')

news_df_human = pd.read_json('../human/news.json')
webnovel_df_human = pd.read_json('../human/webnovel.json')
wiki_df_human = pd.read_json('../human/wiki.json')
```

```python
generated_outputs = pd.concat([
    news_df_generated["output"],
    webnovel_df_generated["output"],
    wiki_df_generated["output"]
], ignore_index=True).tolist()

human_outputs = pd.concat([
    news_df_human["output"],
    webnovel_df_human["output"],
    wiki_df_human["output"]
], ignore_index=True).tolist()

# Combine into dictionary
combined_data = {
    "human": human_outputs,
    "generated": generated_outputs
}

# Save to JSON file
with open("pairwaise_chinese.json", "w", encoding="utf-8") as f:
    json.dump(combined_data, f, ensure_ascii=False, indent=2)
```

```python
def analyse_data(df):
    print(f"Number of rows: {len(df)}")
    print("Maximum length of text:", df['text'].str.len().max())
    print("Minimum length of text:", df['text'].str.len().min())
    print("Average length of text:", df['text'].str.len().mean())
```

```python
analyse_data(combined_df)
```
